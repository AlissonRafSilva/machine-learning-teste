#PREVENINDO CRASHES: TRY / EXCEPT
Na vida real, os dados falham. O usuário digita letra onde era número.

O try/except tenta rodar um código; se der erro, ele captura o erro sem derrubar o programa!

```python
numero_digitado = "A" # Deveria ser um número!

try:
    soma = int(numero_digitado) + 10
    print(soma)
except ValueError:
    print("ERRO DE DADOS: O valor fornecido não é um número válido.")
    print("Ação: Preenchendo com valor padrão (0).")
```

```python
print("\n--- Exemplo 3: Encontrando o Primeiro Número Múltiplo de 3 e 5 ---")

numero_atual = 1

while True:
    if numero_atual % 3 == 0 and numero_atual % 5 == 0:
        print(f"O primeiro número divisível por 3 e 5 é: {numero_atual}")
        break  # Sai do loop quando a condição é satisfeita

    numero_atual += 1
```
